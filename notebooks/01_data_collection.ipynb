# 2. Thực hiện

## 1. Chủ đề dữ liệu (Subject)

### 1.1. Mô tả chủ đề
Dataset này tập trung vào **Mức lương và Xu hướng việc làm trong ngành Khoa học Dữ liệu (Data Science)** trên quy mô toàn cầu từ năm 2020-2025. Bao gồm các thông tin của các vị trị trong lĩnh vực Data Science:
* **job_title:** vị trí công việc
* **employee_residence:** quốc gia đang sinh sống và làm việc 
* **company_location:** quốc gia công ty
* **work_year:** số năm làm việc
* **experience_level:**  cấp độ kinh nghiệm từ cấp độ thực tập sinh (Entry-level) đến chuyên gia (Executive)
* **salary:** mức lương theo năm
* **salary_currency:** đơn vị tiền tệ của mức lương được trả
* **salary_in_usd):** mức lương quy đổi sang đơn vị USD
* **employment_type:** hình thức làm việc Full-time, Part-time
* **remote_ratio:** tỉ lệ công việc làm từ xa (Remote) chiếm bao nhiêu phần trăm trong tổng số công việc

### 1.2. Bối cảnh thực tế
Dataset này nhằm phân tích xu hướng tiền lương, chênh lệch theo địa lý và kinh nghiệm, cũng như mối liên hệ giữa các biến nghề nghiệp và thu nhập — tức là bối cảnh thực tế của thị trường lao động dành cho Data Scientist năm 2025.

## 2. Nguồn dữ liệu (Source)

* **Tên dataset:** Data Science Salaries 2025 
* **Tên nền tảng:** Kaggle
* **Link dataset:** [ Data Science Salaries 2025](https://www.kaggle.com/datasets/arnabchaki/data-science-salaries-2025)
* **Tác giả/Tổ chức:** Tác giả trên Kaggle là **Arnab Chaki**. Tuy nhiên, dữ liệu gốc thường được tổng hợp từ nền tảng **aijobs.net** - một trang web chuyên về tuyển dụng và minh bạch mức lương trong ngành AI/Big Data.
* **Thời gian công bố/thu thập:** Dữ liệu bao gồm các bản ghi từ năm 2020 và được cập nhật liên tục, phiên bản mới nhất được **cập nhật 11/4/2025**.

## 4. Phương thức thu thập dữ liệu 

### 4.1. Phương pháp thu thập
Trên Kaggle không nêu chi tiết phương pháp thu thập, do đó không thể khẳng định chắc chắn phương pháp thu thập. Tuy nhiên, theo đánh giá của chúng em về nguồn dữ liệu, dataset có thể chủ yếu được thu thập thông qua phương pháp:
* **Khảo sát tự báo cáo (Crowdsourcing/Survey)** về vị trí công việc, kinh nghiệm, ... từ cộng đồng người tìm việc trên nền tảng *ai-jobs.net*.
* **Tổng hợp dữ liệu (Aggregation/Scraping)** mức lương, vị trí công ty, ... từ các thông tin công khai trong các tin tuyển dụng được đăng tải trên sàn việc làm của *ai-jobs.net*.

### 4.2. Đối tượng và Thời gian
* **Đối tượng:** Các chuyên gia làm việc trong lĩnh vực Dữ liệu và AI trên toàn thế giới.
* **Thời gian:** Dữ liệu bao trùm khoảng thời gian từ **2020 đến 2025**.

### 4.3. Hạn chế và Thiên kiến (Bias)
* **Thiên kiến tự báo cáo (Self-reporting bias):** Dữ liệu dựa trên sự tự nguyện cung cấp thông tin, nên có thể thiếu chính xác hoặc những người có mức lương cao có xu hướng chia sẻ nhiều hơn.
* **Thiên kiến địa lý:** Dữ liệu có thể tập trung nhiều vào thị trường Mỹ (US) và Châu Âu, trong khi dữ liệu từ các nước đang phát triển có thể ít hơn, dẫn đến việc mức lương trung bình chung có thể cao hơn thực tế tại một số khu vực.

## 5. Lý do lựa chọn bộ dữ liệu 

### 5.1. Sự quan tâm của nhóm
Nhóm chúng em chọn đề tài này vì chủ đề công việc và mức lương trong Data Science thú vị và  rất thực tiễn, có nhiều ứng dụng: so sánh theo quốc gia, theo cấp độ kinh nghiệm, theo ngành/role, dự báo xu hướng lương, hoặc phân tích bất bình đẳng thu nhập trong ngành. Đây là dataset phù hợp cho các bài tập EDA, visualization, trả lời các câu hỏi trong phạm vi đồ án này.

### 5.2. Câu hỏi và Tiềm năng khai thác
Bộ dữ liệu này giúp trả lời các câu hỏi:
1.  **Kinh nghiệm:** Mức lương thay đổi như thế nào giữa các cấp bậc kinh nghiệm?
2.  **Làm việc từ xa:** Công ty đặt ở quốc gia A nhưng tuyển remote ở quốc gia B thì lương có khác biệt gì?
3.  **Vị trí địa lý:** Lương trung vị thay đổi thế nào theo quốc gia?
4.  **Xu hướng theo thời gian:** Mức lương trung bình của ngành có tăng lên qua các năm từ 2020 đến 2025 không? Quốc gia nào có tăng trưởng mức lương nhanh nhất qua từng năm?
